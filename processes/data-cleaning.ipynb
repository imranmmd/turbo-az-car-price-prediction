```python
def priceFormatter(col):
    col = col.replace(' ', '')
    return int(col)
```

```python
df['Price'] = df['Price'].apply(priceFormatter)
```

```python
df['Price'] = df['Price'].astype(float)
df.loc[df['Currency'] == '$', 'Price'] *= 1.7
df.loc[df['Currency'] == 'â‚¬', 'Price'] *= 1.93
```

```python
df.drop('Currency', axis=1, inplace=True)
```

```python
df
```

```python
df[['Year', 'Liters', 'Distance']] = df['Attributes'].str.split(',', expand=True)
```

```python
df.drop('Attributes', axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['Year'] = pd.to_datetime(df['Year'])
```

```python
df['Age'] = 2025 - df['Year'].dt.year
```

```python
df.drop('Year', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
```

```python
df[df.duplicated()].sample(20)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
def cleanLiters(col):
    col = col.replace(' L', '')
    return float(col)
```

```python
df.drop(df[df['Liters'].str.contains('sm3', na=False)].index, inplace=True)
```

```python
df['Liters'] = df['Liters'].apply(cleanLiters)
```

```python
df.info()
```

```python
df
```

```python
def cleanDistance(col):
    col = col.replace('km', '').strip().replace(' ', '')
    return float(col)
```

```python
df['Distance'] = df['Distance'].apply(cleanDistance)
```

```python
df
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
```

```python
df = df[['Car Name', 'Liters', 'Age', 'Distance', 'Price']]
```
